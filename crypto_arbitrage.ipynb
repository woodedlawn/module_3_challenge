#### 1. For each of the three dates, measure the arbitrage spread between the two exchanges by subtracting the lower-priced exchange from the higher-priced one.
```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early = coinbase.loc[early_date, "Close"] - bitstamp.loc[early_date, "Close"]
arbitrage_spread_middle = coinbase.loc[middle_date, "Close"] - bitstamp.loc[middle_date, "Close"]
arbitrage_spread_late = coinbase.loc[late_date, "Close"] - bitstamp.loc[late_date, "Close"]

# Use a conditional statement to generate the summary statistics for each arbitrage_spread DataFrame
print(f"Summary statistics for arbitrage spread on {early_date}")
display(arbitrage_spread_early[arbitrage_spread_early > 0].describe())
print()
print(f"Summary statistics for arbitrage spread on {middle_date}")
display(arbitrage_spread_middle[arbitrage_spread_middle > 0].describe())
print()
print(f"Summary statistics for arbitrage spread on {late_date}")
display(arbitrage_spread_late[arbitrage_spread_late > 0].describe())
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
# For the date early in the dataset, calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early = arbitrage_spread_early[arbitrage_spread_early > 0] / bitstamp.loc[early_date, "Close"]
spread_return_middle = arbitrage_spread_middle[arbitrage_spread_middle > 0] / bitstamp.loc[middle_date, "Close"]
spread_return_late = arbitrage_spread_late[arbitrage_spread_late > 0] / bitstamp.loc[late_date, "Close"]

# Review the spread return DataFrame
print(f"Summary statistics for spread return on {early_date}")
display(spread_return_early.describe())
print()
print(f"Summary statistics for spread return on {middle_date}")
display(spread_return_middle.describe())
print()
print(f"Summary statistics for spread return on {late_date}")
display(spread_return_late.describe())
```

#### 3. For each of the three dates, narrow down your trading opportunities even further. To do so, determine the number of times your trades with positive returns exceed the 1% minimum threshold that you need to cover your costs.

```python
# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs
threshold = .01
profitable_trades_early = spread_return_early[spread_return_early > threshold]
profitable_trades_middle = spread_return_middle[spread_return_middle > threshold]
profitable_trades_late = spread_return_late[spread_return_late > threshold]

# Review the first five profitable trades
print(f"First {5 if len(profitable_trades_early) > 5 else len(profitable_trades_early)} of {len(profitable_trades_early)} profitable trades on {early_date}")
display(profitable_trades_early.dropna().head())
print()
print(f"First {5 if len(profitable_trades_middle) > 5 else len(profitable_trades_middle)} of {len(profitable_trades_middle)} profitable trades on {middle_date}")
display(profitable_trades_middle.dropna().head())
print()
print(f"First {5 if len(profitable_trades_late) > 5 else len(profitable_trades_late)} of {len(profitable_trades_late)} profitable trades on {late_date}")
display(profitable_trades_late.dropna().head())
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

**Question:** After reviewing the profit information across each date from the different time periods, can you identify any patterns or trends?
    
**Answer:** The degree of spread between the closing price of BTC at Bitstamp and Coinbase decreased in Q1 2018. As a result, the number of profitable arbitrage trading opportunities decreased as well. The two charts below show the daily arbitrage profits and cumulative arbitrage profits for Q1. These charts show a significant decrease in arbitrage opportunities in the second half of Q1.

```python
# function to calculate the potential arbitrage profits for a given date
def calculate_profit(date):
    arbitrage_spread = coinbase.loc[date, "Close"] - bitstamp.loc[date, "Close"]
    spread_return = arbitrage_spread[arbitrage_spread > 0] / bitstamp.loc[date, "Close"]
    profitable_trades = spread_return[spread_return > .01]
    profit = profitable_trades * bitstamp.loc[date, "Close"]
    return profit.dropna().sum()

# create a list of date strings for the function above
begin_date = "2018-01-01"
end_date = "2018-03-31"
daterange = pd.date_range(begin_date, end_date)
dates = []
for single_date in daterange:
    dates.append(single_date.strftime("%Y-%m-%d"))

# create a `profits` dataframe to hold the daily potential profits
profits = pd.DataFrame(dates, columns=["date"])
profits["profit"] = profits.apply(lambda row: calculate_profit(row.date), axis = 1)
profits["date"] = pd.to_datetime(profits["date"])
profits.set_index("date", inplace=True)
profits
```

```python
# plot the daily profits
profits.plot(figsize=(20, 10), title=f"Potential Bitcoin Arbitrage Profits from {begin_date} to {end_date}")
```

```python
# plot the cumulative daily profits
cumulative_profits = profits.cumsum()
cumulative_profits.plot(figsize=(20, 10), title=f"Cumulative Bitcoin Arbitrage Profits from {begin_date} to {end_date}")
```
